```python
import pandas as pd
import os
```

```python
gt_test = pd.read_csv('test_GroundTruth2.csv')
gt_test = gt_test[gt_test["none"] == 0]
gt_test = gt_test.drop(columns=["none"])
#save the cleaned dataframe to a new CSV file
gt_test.to_csv('test_GroundTruth2_binary.csv', index=False)
gt_test
```

```python
gt_train = pd.read_csv('train_GroundTruth2.csv')
gt_train = gt_train[gt_train["none"] == 0]
gt_train = gt_train.drop(columns=["none"])
#save the cleaned dataframe to a new CSV file
gt_train.to_csv('train_GroundTruth2_binary.csv', index=False)
gt_train
```

```python
gt_train.sum()
```

```python
image_train_folder = './train/images'
masks_train_folder = './train/masks'
image_test_folder = './test/images'
masks_test_folder = './test/masks'
```

```python
if not os.path.exists('./train_binary'):
    os.makedirs('./train_binary')
    if not os.path.exists('./train_binary/images'):
        os.makedirs('./train_binary/images')
    if not os.path.exists('./train_binary/masks'):
        os.makedirs('./train_binary/masks')
if not os.path.exists('./test_binary'):
    os.makedirs('./test_binary')
    if not os.path.exists('./test_binary/images'):
        os.makedirs('./test_binary/images')
    if not os.path.exists('./test_binary/masks'):
        os.makedirs('./test_binary/masks')
```

```python
# in the new folders, copy the images and masks but only those that are in the cleaned ground truth dataframes
import shutil
for index, row in gt_train.iterrows():
    image_name = row['image_id'] + '.png'
    mask_name = row['image_id'] + '_segmentation.png'
    shutil.copy(os.path.join(image_train_folder, image_name), os.path.join('./train_binary/images', image_name))
    shutil.copy(os.path.join(masks_train_folder, mask_name), os.path.join('./train_binary/masks', mask_name))
for index, row in gt_test.iterrows():
    image_name = row['image_id'] + '.png'
    mask_name = row['image_id'] + '_segmentation.png'
    shutil.copy(os.path.join(image_test_folder, image_name), os.path.join('./test_binary/images', image_name))
    shutil.copy(os.path.join(masks_test_folder, mask_name), os.path.join('./test_binary/masks', mask_name))
```
