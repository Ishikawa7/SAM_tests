```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df_res = pd.read_csv('pca_nb_results.csv')
df_res = df_res.round(3)
df_res
```

```python
df_res[['Dice PCA1', 'Dice PCA2', 'Dice PCA3', 'Dice PCA4', 'Dice PCA5']].describe()
```

```python
df_res[['IoU PCA1', 'IoU PCA2','IoU PCA3', 'IoU PCA4', 'IoU PCA5']].describe()
```

```python
df_res.columns
```

```python
fig = px.violin(df_res[['Dice PCA1', 'Dice PCA2', 'Dice PCA3', 'Dice PCA4', 'Dice PCA5']], points="all", color_discrete_sequence=['darkblue'], title='Dice score distributions', box=True, width=1400, height=700)
fig.update_layout(xaxis_title='Model', yaxis_title='IOU')
fig.write_html('dice_scores_distribution_pca_nb_skincancer_new.html')
fig.write_image('dice_scores_distribution_pca_nb_skincancer_new.png')
fig.show()
```

```python
fig = px.violin(df_res[['IoU PCA1', 'IoU PCA2','IoU PCA3', 'IoU PCA4', 'IoU PCA5']], points="all", color_discrete_sequence=['green'], title='IoU scores distribution', box=True, width=1400, height=700)
fig.update_layout(xaxis_title='Model', yaxis_title='IoU score')
fig.show()
fig.write_html('iou_distribution_pca_nb_skincancer_new.html')
fig.write_image('iou_distribution_pca_nb_skincancer_new.png')
```

```python
fig = px.violin(df_res[['Dice PCA1', 'IoU PCA1']], points="all", color_discrete_sequence=['green'], title='Scores distribution', box=True, width=1400, height=700)
fig.update_layout(xaxis_title='Model', yaxis_title='Score')
fig.show()
fig.write_html('res_pca_nb_scores.html')
fig.write_image('res_pca_nb_scores.png')
```
