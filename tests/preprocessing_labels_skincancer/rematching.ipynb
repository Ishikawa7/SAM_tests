```python
classes = [
    'actinic keratosis', 'melanoma', 'seborrheic keratosis',
    'basal cell carcinoma', 'nevus', 'squamous cell carcinoma',
    'dermatofibroma', 'pigmented benign keratosis', 'vascular lesion'
]
```

```python
folder = "images_classes"
```

```python
# create a dataframe in which each row is an image, one column is the image id (corresponding to the image filename) and the other columns are the classes corresponding to the image folder
dict_df = {
    'image_id': [],
    'class': []
}
for c in classes:
    for image in os.listdir(os.path.join(folder, c)):
        dict_df['image_id'].append(image.split('.')[0])
        dict_df['class'].append(c)
df = pd.DataFrame(dict_df)
df
```
