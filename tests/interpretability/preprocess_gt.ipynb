```python
df_train = pd.read_csv('train_pca_red_skincancer.csv')
# rename first two columns
df_train = df_train.rename(columns={'Unnamed: 0': 'image_id'})
df_train = df_train.drop(columns=['0', '1'])
# one hot encoding column "cluster"
df_train = pd.get_dummies(df_train, columns=['cluster'], dtype=int)
# remove last word "_train" from the image_id column
df_train['image_id'] = df_train['image_id'].str[:-6]
df_train.to_csv('train_umap_skincancer_onehot.csv', index=False)
df_train
```

```python
# do the same for test data
df_test = pd.read_csv('test_pca_red_skincancer.csv')
df_test = df_test.rename(columns={'Unnamed: 0': 'image_id'})
df_test = df_test.drop(columns=['0', '1'])
df_test = pd.get_dummies(df_test, columns=['cluster'], dtype=int)
df_test['image_id'] = df_test['image_id'].str[:-5]
df_test.to_csv('test_umap_skincancer_onehot.csv', index=False)
df_test
```
