```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df_res = pd.read_csv('df_results_skincancer_new_interpretability.csv', index_col=0)
df_res = df_res.round(3)
df_res
```

```python
df_res.columns
```

```python
df_res[['dice_score_ig', 'dice_score_saliency', 'dice_score_deep_lift', 'dice_score_grad_cam']].describe()
```

```python
df_res[['iou_score_ig', 'iou_score_saliency','iou_score_deep_lift', 'iou_score_grad_cam']].describe()
```

```python
df_res.columns
```

```python
fig = px.violin(df_res[['dice_score_ig', 'dice_score_saliency', 'dice_score_deep_lift', 'dice_score_grad_cam']], points="all", color_discrete_sequence=['darkblue'], title='Dice scores distribution', box=True, width=1400, height=700)
fig.update_layout(xaxis_title='Model', yaxis_title='Dice Score')
fig.write_html('dice_score_distribution_interpretabilities.html')
fig.write_image('dice_score_distribution_interpretabilities.png')
fig.show()
```

```python
fig = px.violin(df_res[['iou_score_ig', 'iou_score_saliency','iou_score_deep_lift', 'iou_score_grad_cam']], points="all", color_discrete_sequence=['green'], title='IoU scores distribution', box=True, width=1400, height=700)
fig.update_layout(xaxis_title='Model', yaxis_title='IoU score')
fig.show()
fig.write_html('iou_distribution_interpretabilities.html')
fig.write_image('iou_distribution_interpretabilities.png')
```
